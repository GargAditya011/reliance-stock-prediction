# Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomposition = seasonal_decompose(reliance_data['Open'] , model = 'additive' , period = 22)
fig = decomposition.plot()
fig.set_size_inches(15, 8)
plt.show()
```

# Test Stationarity

```python
from statsmodels.tsa.stattools import adfuller
```

```python
test_result = adfuller(reliance_data['Open'])
print(f'p-value={test_result[1]}')

if(test_result[1]>0.05):
    print("The series is not stationary")
else:
    print("series is stationary")
```

```python
plt.figure(figsize=(15,4))

plt.subplot(1,2,1)
plt.plot(reliance_data['Open'].diff())

plt.subplot(1,2,2)
plt.plot(reliance_data['Open'].diff().diff())

plt.show()
```

```python
from statsmodels.tsa.stattools import acf , pacf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
plot_acf(reliance_data['Open'], ax=ax, alpha = 0.05)
plt.show()

x_acf = pd.DataFrame(acf(reliance_data['Open']))
print(x_acf)
```

```python
plot_pacf(reliance_data['Open'] , lags=20 , alpha = 0.05)
plt.show()
```

# **Arima Model Training**

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model_auto = ARIMA(train['Open'], order=(1, 1, 1))
result_auto = model_auto.fit()
print(result_auto.summary())
```

```python
n = len(test['Open'])
forecast_auto_arima = result_auto.predict(len(train['Open']) , len(train['Open'])+n-1)

start_date = str(train['Open'].iloc[-1:].index[0])

business_days = pd.date_range(start=start_date, periods=n, freq='B')

forecast_auto_arima = pd.DataFrame({'Date': business_days, 'Value': forecast_auto_arima})
forecast_auto_arima.set_index('Date' , drop = True , inplace = True)

plt.figure(figsize = (20,3))
plt.plot(train['Open'] , label = 'Actual_Train' , color = 'red')
plt.plot(test['Open'] , label = 'Actual_Test' , color = 'blue')
plt.plot(forecast_auto_arima , label = 'Forecast' , color = 'green')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
mae = mean_absolute_error(test['Open'], forecast_auto_arima['Value'])

rmse = np.sqrt(mean_squared_error(test['Open'], forecast_auto_arima['Value']))

mape = np.mean(np.abs((test['Open'] - forecast_auto_arima['Value']) / test['Open'])) * 100

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")
print(f"MAPE: {mape}%")
```

# **SARIMAX Model Training**

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model1 = SARIMAX(train['Open'] , order = (2,1,2) , seasonal_order=(1,1,0,12))
result1 = model1.fit()
print(result1.summary())
```

```python
result1.plot_diagnostics(figsize=(16,10))
plt.show()
```

```python
n = len(test['Open'])
test_sarimx_1 = result1.predict(len(train['Open']) , len(train['Open'])+n-1)

start_date = str(train['Open'].iloc[-1:].index[0])

business_days = pd.date_range(start=start_date, periods=n, freq='B')

test_sarimx_1 = pd.DataFrame({'Date': business_days, 'Value': test_sarimx_1})
test_sarimx_1.set_index('Date' , drop = True , inplace = True)

plt.figure(figsize = (20,3))
plt.plot(train['Open'] , label = 'Actual_Train' , color = 'red')
plt.plot(test['Open'] , label = 'Actual_Test' , color = 'blue')
plt.plot(test_sarimx_1 , label = 'Forecast' , color = 'green')
plt.legend()
plt.show()
```

```python
mae = mean_absolute_error(test['Open'], test_sarimx_1['Value'])
rmse = np.sqrt(mean_squared_error(test['Open'], test_sarimx_1['Value']))
mape = np.mean(np.abs((test['Open'] - test_sarimx_1['Value']) / test['Open'])) * 100

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")
print(f"MAPE: {mape}%")
```

```python
plt.figure(figsize = (20,6))
plt.plot(test['Open'] , label = 'Actual_test', color = 'green')
plt.plot(test_sarimx_1['Value'] , label = 'Test_Forecast' , color = 'red')
plt.ylabel("Share Price")
plt.title("Reliance Share price Actual V/S Predicted")
plt.legend()
plt.show()
```
